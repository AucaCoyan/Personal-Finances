# Importo el dolar CCL

```python
# creo un df con el Dolar CCL
dolarccl = pd.read_csv('data/CCL.csv')
dolarccl['Date'] = pd.to_datetime(dolarccl['Date'])
dolarccl.set_index(dolarccl['Date'], inplace=True)
dolarccl.drop(columns='Date', inplace=True)
dolarccl
```

```python
# Agrego la columna valor del CCL en el día de la operación
df['CCL del dia'] = 0.0   # np.arange(0, 353, 1).astype(float)    # valor placeholder. Un float porque después lo tengo que reemplazar por otro float

# coloco el dolar segun el df dolarccl
for operacion in df.index:  # por cada operacion (= indice de df) en el indice de df hago:
    columna = 'Liquida'     # digo que la columna donde voy a sacar los datos es 'Liquida', a futuro lo puedo cambiar por 'Operado'

    # Busco el CCL de la tabla dolarccl, de acuerdo a la fecha
    ccl_operacion = dolarccl.loc[
                                df.at[operacion, columna]
                                ].round(3)
    # print(ccl_operacion)    # para verificacion a ver si está importando bien el CCL

    df.at[operacion, 'CCL del dia'] = ccl_operacion               # y la igualo ccl_operacion a la celda de la operacion de este ciclo

# Chequeo que no haya NaN en CCL
missing_CCL = df['CCL del dia'].isnull().sum()
missing_CCL
```

```python
# Importe USD  == Lo que salió la operación en USD
df['ImporteUSD'] = (df['Importe'] / df['CCL del dia']).round(2)
# Saldo USD == lo que tengo en la cuenta saldo en ARS, pasado a USD
df['SaldoUSD'] = (df['Saldo'] / df['CCL del dia']).round(2)
```
